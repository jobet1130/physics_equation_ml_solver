```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import os 

sns.set_style("whitegrid")

DATA_PATH = "../../data/processed/newtons_second_law_cleaned_dataset.csv"

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Cleaned dataset not found at: {DATA_PATH}")
```

```python
df = pd.read_csv(DATA_PATH)
print(f"Loaded cleaned dataset: {df.shape[0]} rows x {df.shape[1]} columns")
df.head()
```

```python
df.describe().T
```

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(df["Mass_kg"], bins=30, kde=True, color="skyblue")
plt.title("Mass Distribution (kg)")

plt.subplot(1, 3, 2)
sns.histplot(df["Acceleration_ms2"], bins=30, kde=True, color="salmon")
plt.title("Acceleration Distribution (m/s)")

plt.subplot(1, 3, 3)
sns.histplot(df["Force_N"], bins=30, kde=True, color="mediumseagreen")
plt.title("Force Distribution (N)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x="Mass_kg", y="Force_N", alpha=0.6, color="steelblue")
plt.title("Force vs Mass")

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x="Acceleration_ms2", y="Force_N", alpha=0.6, color="darkorange")
plt.title("Force vs Acceleration")

plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(16, 4))

for i, col in enumerate(["Mass_kg", "Acceleration_ms2", "Force_N"]):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x=df[col], color="lightgray")
    plt.title(f"Boxplot: {col}")

plt.tight_layout()
plt.show()
```
