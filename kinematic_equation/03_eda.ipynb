```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import os 

sns.set_style("whitegrid")

DATA_PATH = "../../data/processed/kinematic_equation_cleaned_dataset.csv"

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Cleaned dataset not found at: {DATA_PATH}")
```

```python
df = pd.read_csv(DATA_PATH)
print(f"Loaded cleaned dataset: {df.shape[0]} rows x {df.shape[1]} columns")
df.head()
```

```python
df.describe().T
```

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(df["Acceleration_ms2"], bins=30, kde=True, color="skyblue")
plt.title("Acceleration Distribution (m/s²)")

plt.subplot(1, 3, 2)
sns.histplot(df["Time_s"], bins=30, kde=True, color="salmon")
plt.title("Time Distribution (s)")

plt.subplot(1, 3, 3)
sns.histplot(df["Displacement_m"], bins=30, kde=True, color="mediumseagreen")
plt.title("Displacement Distribution (m)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x="Acceleration_ms2", y="Displacement_m", alpha=0.6, color="steelblue")
plt.title("Displacement vs Acceleration")

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x="Time_s", y="Displacement_m", alpha=0.6, color="darkorange")
plt.title("Displacement vs Time")

plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(16, 4))

for i, col in enumerate(["Acceleration_ms2", "Time_s", "Displacement_m"]):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x=df[col], color="lightgray")
    plt.title(f"Boxplot: {col}")

plt.tight_layout()
plt.show()
```
